## Introduction

In this notebook, I'll analyze a dataset containing demographics and passenger information from 891 of the 2224 passengers and crew on board the Titanic. Although there was some element of luck involved in surviving the sinking, we can assume that some groups of people were more likely to survive than others. 

In this analysis, I'll look for patterns that describe how the population who survived the sinking differed from the population who didn't. Based on what we know about the accident, that women and children were prioritized and probably class influenced who were evacuated first, here are the quesitons I'm going to explore:



### Exploratory Questions
_____

1. **How does survival rate change by class? Are wealthier individuals (1st class) more likely to survive than the poor (3rd class)?**
2. **How does survival rate change by gender? Does gender affect the chances of surviving?**
3. **How does age impact survival rates? Are older people more likely to die?**
_____


## Importing CSV data into a Pandas DataFrame

```python
titanic_filepath = 'titanic_data.csv'
raw_titanic_df = pd.read_csv(titanic_filepath)
raw_titanic_df.head(3)
```

## Treating and Cleaning data

```python
# Getting summary info from raw dataframe
raw_titanic_df.info()
```

```python
# Dropping unnecessary (that won't be used in the analysis) columns from raw dataframe
clean_titanic_df = raw_titanic_df.drop(['PassengerId','Ticket', 'Cabin'], axis=1)
# Getting summary info from cleaned dataframe
clean_titanic_df.info()
```

```python
# Filtering data set to only those passangers with full features infromation
full_titanic_df = clean_titanic_df.dropna()
# Getting summary info from full dataframe
full_titanic_df.info()
```

```python
#Creating a new treated dataframe
pd.options.mode.chained_assignment = None
titanic_df = full_titanic_df
titanic_df['Male'] = (titanic_df['Sex'] == 'male')
titanic_df['Female'] = (titanic_df['Sex'] == 'female')

titanic_df['Baby'] =  (titanic_df.Age <= 2)
titanic_df['Child'] = ((titanic_df.Age <= 12) & (titanic_df.Age > 2))
titanic_df['Young'] = ((titanic_df.Age < 20) & (titanic_df.Age > 12))
titanic_df['Adult'] = ((titanic_df.Age < 60) & (titanic_df.Age >= 20))
titanic_df['Old'] =  (titanic_df.Age >= 60)

titanic_df['Family?'] =  ((titanic_df.SibSp > 0) | (titanic_df.Parch > 0))

titanic_df['1st_Class'] =  (titanic_df.Pclass == 1)
titanic_df['2nd_Class'] =  (titanic_df.Pclass == 2)
titanic_df['3rd_Class'] =  (titanic_df.Pclass == 3)

titanic_df['AgeGroup'] = np.where(titanic_df.Baby == True, 1, 0)
titanic_df['AgeGroup'] = np.where(titanic_df.Child == True, 2, titanic_df['AgeGroup'])
titanic_df['AgeGroup'] = np.where(titanic_df.Young == True, 3, titanic_df['AgeGroup'])
titanic_df['AgeGroup'] = np.where(titanic_df.Adult == True, 4, titanic_df['AgeGroup'])
titanic_df['AgeGroup'] = np.where(titanic_df.Old == True, 5, titanic_df['AgeGroup'])

titanic_df.head(2)
```

## Exploring the Questions

### 1. How does survival rate change by class?

#### Analyzing Observations

If we calculate the survival rate and plot the number of people who survived and died for each class, from the sample we get that being from a higher class leads to a greater survival rate. 

As calculated below for the sample, the 1st class survival rate is 65.2%, while the 2nd class rate is 48.0% and the 3rd's is 23.9%.

```python
# Calculating survival rates for each class
print 'Survival Rate for each class is'
print '- 1st Class: %1.3f' %(titanic_df.Survived[titanic_df.Pclass == 1].mean())
print '- 2nd Class: %1.3f' %(titanic_df.Survived[titanic_df.Pclass == 2].mean())
print '- 3rd Class: %1.3f' %(titanic_df.Survived[titanic_df.Pclass == 3].mean())
```

```python
# Plotting survival rate by class
class_survival_rate_df = titanic_df.groupby('Pclass').mean().reset_index()
plt.scatter(class_survival_rate_df.Pclass, class_survival_rate_df.Survived,alpha= 0.6, s=2*titanic_df.groupby('Pclass').count()['Survived'].values)
plt.xlabel("Class")
plt.ylabel("Survial Rate")
plt.title("Survial Rate by Class (Size = # passengers in the class)")
plt.xticks(class_survival_rate_df.Pclass, ['1st', '2nd', '3rd'])
```

*Are these differences significant?*

From the p-values calculated below, even at a confidence level of 99%, these differences are significant. So, we can affirm that in Titatic the survival rate changed by class, with wealthier individuals (1st class) being more likely to survive than the poor (3rd class).

```python
# Calculating t Statistics for each difference in survival rate among classes
print 't-Test results for differences in survival rates:'
print ''
print '- 1st and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df.Pclass == 1], titanic_df.Survived[titanic_df.Pclass == 3], equal_var=False)[1])
print '- 1st and 2nd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df.Pclass == 1], titanic_df.Survived[titanic_df.Pclass == 2], equal_var=False)[1])
print '- 2nd and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df.Pclass == 2], titanic_df.Survived[titanic_df.Pclass == 3], equal_var=False)[1])
```

*Going Deeper*

Even though the results are significant if we look at classes separately, they might be due to other variables with strong correlation to survival and which may have different distributions across classes. One variable that likely causes the difference is gender. So, it's worth taking a look at survival rates across classes for different genders, and analyzing how class influences survival for each gender. That's what we do below:

```python
# Plotting survival rate by class and gender
class_survival_rate_males_df = titanic_df[titanic_df.Sex == 'male'].groupby('Pclass').mean().reset_index()
plt.scatter(class_survival_rate_males_df.Pclass, class_survival_rate_males_df.Survived,alpha= 0.6, s=2*titanic_df[titanic_df.Sex == 'male'].groupby('Pclass').count()['Survived'].values, color='b')
class_survival_rate_females_df = titanic_df[titanic_df.Sex == 'female'].groupby('Pclass').mean().reset_index()
plt.scatter(class_survival_rate_females_df.Pclass, class_survival_rate_females_df.Survived,alpha= 0.6, s=2*titanic_df[titanic_df.Sex == 'female'].groupby('Pclass').count()['Survived'].values, color='magenta')

plt.xlabel("Class")
plt.ylabel("Survial Rate")
plt.title("Survial Rate by Class and Gender (Size = # of passangers)")
plt.xticks(class_survival_rate_males_df.Pclass, ['1st', '2nd', '3rd'])
plt.legend(('Males', 'Females'),loc='best')
```

```python
# Calculating survival rates for each gender and class group
print 'Survival Rate for each class and gender is'
print ''
print '- 1st Class Males: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'male')].mean())
print '- 2nd Class Males: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'male')].mean())
print '- 3rd Class Males: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'male')].mean())
print ''
print '- 1st Class Females: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'female')].mean())
print '- 2nd Class Females: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'female')].mean())
print '- 3rd Class Females: %1.3f' %(titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'female')].mean())
```

From the charts and data above, we can notice that survival rates change a lot for different genders - a fact that we'll explore in the next question. And though we can notice that within each gender survival rate increases for higher classes, their differences are not as signigificant as if we looked at classes alone. That indicates that what we suspected is true: the different distribution of gender within each class accounts for part of the differences in survival rates.

But how significant are these differences? Next, we explore it.

#### Statistical Tests

```python
# Calculating t Statistics for each difference in survival rate among classes and gender
print 't-Test results for differences in survival rates:'
print ''
print '- Males in 1st and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'male')], titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'male')], equal_var=False)[1])
print '- Males in 1st and 2nd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'male')], titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'male')], equal_var=False)[1])
print '- Males in 2nd and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'male')], titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'male')], equal_var=False)[1])
print ''
print '- Females in 1st and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'female')], titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'female')], equal_var=False)[1])
print '- Females in 1st and 2nd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 1) & (titanic_df.Sex == 'female')], titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'female')], equal_var=False)[1])
print '- Females in 2nd and 3rd classes, p=%f' %(stats.ttest_ind(titanic_df.Survived[(titanic_df.Pclass == 2) & (titanic_df.Sex == 'female')], titanic_df.Survived[(titanic_df.Pclass == 3) & (titanic_df.Sex == 'female')], equal_var=False)[1])
```

#### Conclusion

With the t-Tests calculated above, we can conclude that:
- For men, the differences in survival rate from the 1st class to the 2nd and 3rd classes are significant. But the difference between the 2nd and the 3rd isn't significant. So we can say that for men, being from the 1st class increased chances of survival when compared to being from the 2nd or 3rd classes, which had about the same survival rate.
- For women, the differences in survival rate from the 3rd class to the 1st and 2nd classes are significant. But the difference between the 1st and the 2nd isn't significant. So, for women, being from either the 1st or 2nd class didn't change the chance of survival significantly, but being from the 3rd class made them half as likely to survive.

### 2. How does survival rate change by gender?

#### Analyzing Observations

In this sample, following what we expected from knowing that women were evacuated first, we see a strong difference in survival rates based on gender, indicating that women were more likely to survive than men.

As shown in the calculations below, women's survival rate was 75.3% and men's, 20.5%.

```python
# Calculating survival rates for each gender
print 'Survival Rate for each gender is'
print '- Males: %1.3f' %(titanic_df.Survived[titanic_df.Sex == 'male'].mean())
print '- Females: %1.3f' %(titanic_df.Survived[titanic_df.Sex == 'female'].mean())
```

```python
# Plotting survival rate by gender
gender_survival_rate_df = titanic_df.groupby('Sex').mean().reset_index()
plot = gender_survival_rate_df.Survived.plot(kind='bar', color='mb', alpha=0.5)
plt.xlabel("Gender")
plt.ylabel("Survial Rate")
plt.title("Survial Rate by Gender")
plt.xticks([0, 1], ['Females', 'Males'],rotation='horizontal')
```

#### Statistical Test

```python
# Calculating t Statistics for gender survival rates difference
print 't-Test result for the difference in survival rates:'
print ''
print 'for males and females, t=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df.Sex == 'female'], titanic_df.Survived[titanic_df.Sex == 'male'], equal_var=False)[0])
print 'and p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df.Sex == 'female'], titanic_df.Survived[titanic_df.Sex == 'male'], equal_var=False)[1])
```

#### Conclusion

From the t-Test results above, we can conclude that the difference in survival rates between women and men are significant. So, in fact women were more likely to survive than men.

#### Analyzing Observations

From the results and charts above we can notice a general trend for survivers being younger than the ones who died.

In the first chart, where we plot the survival rates for each of the age groups - and where the area of the bubles represent the group size, we can see a tendency of decreased survival rate for older people. Here, we can also notice that the greatest age group aboard were adults.

Finally, calculating the survival rates for each age group, we get a clear trend: the older the group, the lower is the survival rate. But are these differences significant? We explore this question with the t-Test performed below:

#### Statistical Tests

```python
# Calculating t Statistics for each age group survival rate differences
print '- Babies and Children, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 1], titanic_df.Survived[titanic_df['AgeGroup'] == 2], equal_var=False)[1])
print '- Babies and Young, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 1], titanic_df.Survived[titanic_df['AgeGroup'] == 3], equal_var=False)[1])
print '- Babies and Adults, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 1], titanic_df.Survived[titanic_df['AgeGroup'] == 4], equal_var=False)[1])
print '- Babies and Old, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 1], titanic_df.Survived[titanic_df['AgeGroup'] == 5], equal_var=False)[1])
print ''
print '- Children and Young, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 2], titanic_df.Survived[titanic_df['AgeGroup'] == 3], equal_var=False)[1])
print '- Children and Adults, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 2], titanic_df.Survived[titanic_df['AgeGroup'] == 4], equal_var=False)[1])
print '- Children and Old, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 2], titanic_df.Survived[titanic_df['AgeGroup'] == 5], equal_var=False)[1])
print ''
print '- Young and Adults, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 3], titanic_df.Survived[titanic_df['AgeGroup'] == 4], equal_var=False)[1])
print '- Young and Old, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 3], titanic_df.Survived[titanic_df['AgeGroup'] == 5], equal_var=False)[1])
print ''
print '- Adults and Old, p=%f' %(stats.ttest_ind(titanic_df.Survived[titanic_df['AgeGroup'] == 4], titanic_df.Survived[titanic_df['AgeGroup'] == 5], equal_var=False)[1])
```
