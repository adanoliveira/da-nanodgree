# Introduction

In 2000, Enron was one of the largest companies in the United States. By 2002, it had collapsed into bankruptcy due to widespread corporate fraud. In the resulting Federal investigation, a significant amount of typically confidential information entered into the public record, including tens of thousands of emails and detailed financial data for top executives. 

In this project, we'll build a machine learning model to identify people who were somehow involved in the fraud, based on financial and email data made public as a result of the Enron scandal. Since machine learning is a great tool reaching conclusions and making predictins from patterns and trends in data, it's the right tool for the job.

Our dataframe was created by combining this data with a hand-generated list of persons of interest (POI) in the fraud case, which means individuals who were indicted, reached a settlement or plea deal with the government, or testified in exchange for prosecution immunity.

Our goal is to build a classifier that can predict, by looking at financial and email data, if a given individual is a POI or not, with precision and recall being both at least 0.3.

## 2.1 Inspecting for Correlations

To get a sense of the importace that each raw feature has in predicting POIs, let's see how the features correlate our target, and how they correlate among themselves.

```python
numeric_fatures = raw_data_df.select_dtypes(include=['float64']).columns.tolist()
correl_df = pd.DataFrame(columns=['feature', 'correlation', 'pvalue'])
data_df = raw_data_df.copy()

for feature in numeric_fatures:
    correl = round(float(stats.pointbiserialr(data_df.poi, data_df[feature].fillna(0))[0]),2)
    pvalue = round(float(stats.pointbiserialr(data_df.poi, data_df[feature].fillna(0))[1]),3)
    aux_df = pd.DataFrame(data=[[feature, correl, pvalue]], columns=['feature', 'correlation', 'pvalue'])
    correl_df = correl_df.append(aux_df).reset_index(drop=True)

correl_df['abs_correl'] = correl_df.correlation.apply(lambda x: abs(x))
print 'Point biserial correlation coef. of numeric features with target: \n'
print correl_df.sort_values(by='abs_correl', ascending=False).reset_index(drop=True)[['feature', 'correlation', 'pvalue']]
```

## 2.2 Creating new features

### Fraction of emails to and from POIs

Comparing only the absolute number of email sent to and from POIs might be useful for or prediction, but since the number of emails a person sends can very, it tends to lead to assymetric comparisons. I suspect that a better indicator of relationship with POIs would be the percentage of the total emails of a person that were relate to a POI. And that's the first feature we're going to create. Definitions:

- fraction_from_poi_to_this_person = from_poi_to_this_person / to_messages
- fraction_from_this_person_to_poi = from_this_person_to_poi / from_messages


```python
raw_data_df['fraction_from_poi_to_this_person'] = raw_data_df.from_poi_to_this_person / raw_data_df.to_messages
raw_data_df['fraction_from_this_person_to_poi'] = raw_data_df.from_this_person_to_poi / raw_data_df.from_messages
raw_data_df[['fraction_from_poi_to_this_person', 'fraction_from_this_person_to_poi']].hist(figsize=(10,5), bins=20);
```

### Fraction of total compensation

The fraction of a given type of compensation of the total compensasion given to a person can also add information to our raw feature, appart from the absolute values. For instance, by comparing the fraction of exercised stock options from the total stock value, our model could probably distinguish POIs.

```python
#As deferred values are negative, I chose to sum them to the aggregate values in order to compute the fractions
raw_data_df['aux_total_payments'] = raw_data_df.total_payments - raw_data_df.deferred_income
raw_data_df['aux_total_stock_value'] = raw_data_df.total_stock_value - raw_data_df.restricted_stock_deferred 
for payment_feature in total_payments:
    raw_data_df[payment_feature + '_fraction_from_payments'] = raw_data_df[payment_feature] / raw_data_df.aux_total_payments
for stock_feature in total_stock_value:
    raw_data_df[stock_feature + '_fraction_from_stock'] = raw_data_df[stock_feature] / raw_data_df.aux_total_stock_value
raw_data_df['payments_fraction_from_compensation'] = raw_data_df.total_payments / (raw_data_df.total_payments + raw_data_df.total_stock_value)
raw_data_df['stock_fraction_from_compensation'] = raw_data_df.total_stock_value / (raw_data_df.total_payments + raw_data_df.total_stock_value)
raw_data_df.drop(['aux_total_payments', 'aux_total_stock_value'], axis=1, inplace=True)
```

### Fraction of emails predicted as from POI

Our dataset already had email related features, but they reflected only the patterns in the number of emails sent to or received from a POI, as well the overall number of emails sent. Assuming that the content of emails might also provide relevant distinctive information, I decided to include it in our dataset. However, given the high dimensionality of text, including the actual content to our dataset would make it very big and probably distort the impact of other features. So I decided to use the output of a text classifier as a single feature.

From the [Enron Email Dataset](https://www.cs.cmu.edu/~./enron/), which contains email messages from enron employees, I got emails associated to people in our dataset to create a Multinomial Naive Bayes classifier that predicts if a given email is from a POI or non-POI, base on its contents. For every person in our dataset (who had emails in the enron corpus), that classifier was then used to predict whether each of their emails was from a POI or not. And the fraction of emails predicted as from POI became the feature of choice. 

The steps used to crate this feature are outlined below:

- Parsed all emails from people in our dataset
- Trained a Multinomial Naive Bayes classifier using 90% of the emails
- Tested and measured results
- Used trained model to classify each email of every person
- Computed the fraction of emails classified as from POI to use as our feature

# 3. Picking and Tuning an Algorithm

In this section several algorithms will be tested with the chosen feature set (`'exercised_stock_options','fraction_from_this_person_to_poi', 'bonus_fraction_from_payments'`) and several parameter combinations, using GridSearchCV and startified shfflesplit cross validation with 1000 folds. First, we'll tune the parameters of each algorithm to find the best combination. Then, we'll compare the best tuned models of each algorithm to find our final choice, based mainly on f1-score (considering that both precision and recall are above 0.3). The performance of the models will be tested using cross validation to guarantee that it will reflect the the model's ability to generalize to new data, which is precisely what it will encounter once we put it to use. If we measure it's performance using the same data used for training it, we'd be overestimating its power. 

Each of the following algorithms will be tested and have and their main parameters tuned in this step.

- Gaussian Naive Bayes
- Support Vector Classifier
- Decistion Tree
- k Nearest Neighbor
- AdaBoost
- Random Forest

Parameters are arguments that you pass when you create your classifier in order to change how it calculates the decision boundary used to make predictions. They don't completely change the inner workings of the algorithm, but greatly impact its performance. And given the wide range of possible their combinations, it's good to always test sistematically different parameter tunnings to find an optimal combination that enhance f-score by several points, when compared to the out-of-the-box algorithm.

From the stand-alone algorithms, we can see that Gaussian Naive Bayes surprisingly yielded the best F1-score, even not having any parameter tuned. And when using ensembles, Random Forest generated a slightly better F1-score. However, it wasnt'the best in all metrics. In accuracy, which is the simplest performance metric, expressing the fraction of datapoints classified correctly over all data points, k Nearest Neighbors is the best. But accuracy most times is misleading, mainly when used in problems with unbalanced classes, like this one. Just to ilustrate this point, as almost 88% of people in our dataset aren't POIs, if our model classified all data points as non-POIs it would have an accuracy of 88%, even though it missed all POIs. 

A better way to measure performance is using precision and recall. Precision is the probability of an item actually being of a given class, once the algorithm predicted the item as being from that class, and recall is the probability of an algorithm classifying an item in a given class, provided that the item is from that class. Both when used together reflect the power of a model to make correct predictions. The model that had the highest precision was kNN (0.6287), i.e., it's the one with the highest probability of having a person classified as POI actually being a POI. And the model with highest recall was the decision tree (0.8025), i.e., it's the one with the highest chance of classifying an actual POI as being a POI. But they performed relatively poorer in the other metric, either recall (kNN) or precision (DT). In other words, while the decision tree had a higher change of classifying an actual POI as being a POI, it did that at the expense of precision, classifying a bunch of non-POIs as being POIs. 

Depending on the use we make of the model, we might prioritize precision more than recall, or the opposite. Let's say that we would apply this model to select people for an increased level of scrutiny in our investigation. In that case, recall would be more desirable and we would choose the decision tree, because the consequences for our false positives wouldn't be considerably negative. On the other hand, if I used this model to actually prosecute people predicted as being POIs, then precision would become more important, because having a false positive would create very negative consequences, and we'd go with the kNN. However, since I don't have a specific use case in mind, I prefer to go with a balanced model in terms of these two metrics, which is expressed by the F1-score. So, I choose the model with the highest F1-score, in this case, the model with the optimized version of a Random Forest algorithm, which had a F1-score of 0.574.
